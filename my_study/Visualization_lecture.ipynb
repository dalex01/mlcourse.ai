```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/telecom_churn.csv')
```

```python
df.head()
```

```python
df.shape
```

## Whole dataset visualization

```python
df['International plan'] = df['International plan'].map({'Yes': 1, 'No': 0})
df['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1, 'No': 0})
df['Churn'] = df['Churn'].astype('int')
```

```python
df.head()
```

```python
plt.rcParams['figure.figsize'] = (16, 12)
df.drop(['State'], axis=1).hist();
```

```python
sns.heatmap(df.corr());
```

Drop redundant (with good correlation) features

```python
df.drop([feat_name for feat_name in df.columns if 'charge' in feat_name], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.manifold import TSNE
```

```python
df.shape
```

```python
state = df['State']
df.drop('State', axis=1, inplace=True)
```

```python
df.head()
```

```python
tsne=TSNE(random_state=17)
```

```python
%%time
X_repr=tsne.fit_transform(df)
```

```python
X_repr.shape
```

```python
#X_reps_churn = tsne.fit_transform(df[df['Churn'] == 1])
plt.rcParams['figure.figsize'] = (8,6)
plt.scatter(X_repr[df['Churn']==1,0], 
            X_repr[df['Churn']==1,1], alpha=.5, c='blue', label='churn');
plt.scatter(X_repr[df['Churn']==0,0], 
            X_repr[df['Churn']==0,1], alpha=.5, c='orange', label='loyal');
plt.xlabel('tsne axis #1');
plt.ylabel('tsne axis #2');
plt.legend()
plt.title('tSNE representation');
plt.savefig('churn.tsne.png', dpi=300);
```
